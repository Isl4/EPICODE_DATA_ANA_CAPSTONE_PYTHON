### step 3: convert order date column in datetime 

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True, errors="coerce")
df["Ship Date"] = pd.to_datetime(df["Ship Date"], dayfirst=True, errors="coerce")
df
```

### Step 4: sales by year and create visuals 

```python
Sales_by_Year=df.groupby(df['Order Date'].dt.year)['Sales'].sum().reset_index()
Sales_by_Year.columns=['Year','Total Sales']

Sales_by_Year
```

```python
#visulas 

plt.figure(figsize=(10,5))
plt.bar(Sales_by_Year['Year'], Sales_by_Year['Total Sales'], color='skyblue')
plt.title('Total Sales by Year')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.xticks(Sales_by_Year['Year'])
plt.show()
```

### Step 5: dentifying high‑sales / low‑profit sub‑categories

```python
subcat=df.groupby(df['Sub-Category'])[['Sales','Profit']].sum().reset_index()

# calcualte the median of sales and profit
Sales_median= subcat['Sales'].median()
Profit_median= subcat['Profit'].median()

# filter high sales&low profit
fil_subcat=subcat[(subcat['Sales']> Sales_median)& (subcat['Profit']<Profit_median)]

fil_subcat
```

```python
# just to check
subcat=df.groupby(df['Sub-Category'])[['Sales','Profit']].sum().reset_index()
subcat
```

```python
#visual for sales and profit together 
subcat_sorted = subcat.sort_values('Sales', ascending=False)

plt.figure(figsize=(14,7))

x = range(len(subcat_sorted))

plt.bar(x, subcat_sorted['Sales'], label='Sales')
plt.bar(x, subcat_sorted['Profit'], label='Profit')

plt.xticks(x, subcat_sorted['Sub-Category'], rotation=90)
plt.legend()
plt.title('Sales and Profit by Sub-Category')

plt.show()
```

```python
# still with subcategories, now sales by segment 

tables = df[df['Sub-Category'] == 'Tables']

segment_sales = tables.groupby('Segment')[['Sales','Profit']].sum().reset_index()

print(segment_sales)
```

### Step 6:  Check the relation and sales

```python
Disc_Sal=tables.groupby('Segment').agg({'Discount':'mean','Sales':'sum','Profit':'sum'}).reset_index()

Disc_Sal['Discount']=(Disc_Sal['Discount']*100).round(2)

Disc_Sal
```

```python
#compare discounts value across segments --> bar chart
plt.figure(figsize=(8,5))

sns.barplot(data=Disc_Sal, x='Segment', y='Discount')

plt.gca().yaxis.set_major_formatter(lambda x, _: f'{x:.0%}')

plt.title('Average Discount by Segment (Tables)')
plt.show()
```

```python
tables = df[df['Sub-Category']=='Tables'].copy()
```

```python
# heigh discount leads to low profit: 
#black line shows : As Discount increases, Profit tends to decrease.
#Red dashed horizontal line at Profit = 0
#Orders above this line made positive profit.
#Orders below this line caused a loss.

plt.figure(figsize=(8,5))
sns.scatterplot(data=tables, x='Discount', y='Profit', hue='Segment')
sns.regplot(data=tables, x='Discount', y='Profit', scatter=False, color='black')
plt.axhline(0, color='red', linestyle='--', label='Break-even Profit')
plt.title('Profit vs Discount (Tables)')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.show()
```

### Step 8: Shipping & Delivery Performance Analysis

```python
#Delivery Time Analysis
df['Delivery Days']=(df['Ship Date']-df['Order Date']).dt.days
df['Delivery Days'].describe()

#result : negative values 
```

```python
df_clean = df[(df["Delivery Days"] >= 0) & (df["Delivery Days"] <= 30)]
df_clean["Delivery Days"].describe()

## Result : Typical delivery time = 2–4 days
```

```python
df_clean.groupby("Ship Mode")["Delivery Days"].mean()
```

```python
df_clean.groupby("Order Priority")["Delivery Days"].mean()
```

```python
df_clean[["Delivery Days","Profit"]].corr()

##Result There is no relationship between delivery speed and profit
```

```python
#Order Priority vs Delivery Speed Analysis
df_clean.groupby("Order Priority")["Delivery Days"].mean()



#Result :
# Priority is Respected
# ritical Orders (2.47 days)-->1. Fastest fulfillment 2. Likely expedited handling
#Low Orders --> (6.22 days) 1. Significantly slower  2.Likely shipped using economical methods
```

***The company effectively prioritizes urgent orders, reducing delivery time for critical requests by nearly 60% compared to low-priority orders. However, delivery speed has no measurable impact on profitability, suggesting that shipping performance is a customer service lever rather than a revenue driver.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

sns.barplot(
    data=df_clean,
    x="Ship Mode",
    y="Delivery Days",
    estimator="mean"
)

plt.title("Average Delivery Days by Ship Mode")
plt.xticks(rotation=45)
plt.ylabel("Avg Delivery Days")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df_clean,
    x="Order Priority",
    y="Delivery Days"
)

plt.title("Delivery Days Distribution by Order Priority")
plt.ylabel("Delivery Days")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df_clean,
    x="Delivery Days",
    y="Profit",
    alpha=0.6
)

plt.title("Delivery Days vs Profit")
plt.show()
```

### Step 9: 4️⃣ Profit Margin Analysis

```python
df['Profit Margin']= df['Profit']/df['Sales']
df['Profit Margin']
```

```python
df['Profit Margin %'] = np.where(
    df['Sales'] != 0,
    (df['Profit'] / df['Sales']) * 100,
    0
)
df['Profit Margin'].describe()
```

```python
df[df['Profit Margin'] < -1]
```

```python
df.groupby('Category')['Profit Margin'].mean().sort_values()
```

```python
df.groupby('Sub-Category')['Profit Margin'].mean().sort_values()
```

```python
sns.scatterplot(x='Discount', y='Profit Margin', data=df)
```

***The overall average profit margin is 15.1%, however the distribution is negatively skewed due to extreme loss-making transactions. Some orders show margins below -100%, indicating that aggressive discounting or high costs may be driving losses. Strategic discount optimization is recommended.
