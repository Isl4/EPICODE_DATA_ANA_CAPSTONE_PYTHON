### step 3: convert order date column in datetime 

```python
df['Order Date']= pd.to_datetime(df['Order Date'],format='mixed', dayfirst=False, errors='coerce')
df['Ship Date']= pd.to_datetime(df['Ship Date'],format='mixed', dayfirst=False, errors='coerce')
df
```

### Step 4: sales by year and create visuals 

```python
Sales_by_Year=df.groupby(df['Order Date'].dt.year)['Sales'].sum().reset_index()
Sales_by_Year.columns=['Year','Total Sales']

Sales_by_Year
```

```python
#visulas 

plt.figure(figsize=(10,5))
plt.bar(Sales_by_Year['Year'], Sales_by_Year['Total Sales'], color='skyblue')
plt.title('Total Sales by Year')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.xticks(Sales_by_Year['Year'])
plt.show()
```

### Step 5: dentifying high‑sales / low‑profit sub‑categories

```python
subcat=df.groupby(df['Sub-Category'])[['Sales','Profit']].sum().reset_index()

# calcualte the median of sales and profit
Sales_median= subcat['Sales'].median()
Profit_median= subcat['Profit'].median()

# filter high sales&low profit
fil_subcat=subcat[(subcat['Sales']> Sales_median)& (subcat['Profit']<Profit_median)]

fil_subcat
```

```python
# just to check
subcat=df.groupby(df['Sub-Category'])[['Sales','Profit']].sum().reset_index()
subcat
```

```python
#visual for sales and profit together 
subcat_sorted = subcat.sort_values('Sales', ascending=False)

plt.figure(figsize=(14,7))

x = range(len(subcat_sorted))

plt.bar(x, subcat_sorted['Sales'], label='Sales')
plt.bar(x, subcat_sorted['Profit'], label='Profit')

plt.xticks(x, subcat_sorted['Sub-Category'], rotation=90)
plt.legend()
plt.title('Sales and Profit by Sub-Category')

plt.show()
```

```python
# still with subcategories, now sales by segment 

tables = df[df['Sub-Category'] == 'Tables']

segment_sales = tables.groupby('Segment')[['Sales','Profit']].sum().reset_index()

print(segment_sales)
```

### Step 6:  Check the relation and sales

```python
Disc_Sal=tables.groupby('Segment').agg({'Discount':'mean','Sales':'sum','Profit':'sum'}).reset_index()

Disc_Sal['Discount']=(Disc_Sal['Discount']*100).round(2)

Disc_Sal
```

```python
#compare discounts value across segments --> bar chart
plt.figure(figsize=(8,5))

sns.barplot(data=Disc_Sal, x='Segment', y='Discount')

plt.gca().yaxis.set_major_formatter(lambda x, _: f'{x:.0%}')

plt.title('Average Discount by Segment (Tables)')
plt.show()
```

```python
tables = df[df['Sub-Category']=='Tables'].copy()
```

```python
# heigh discount leads to low profit: 
#black line shows : As Discount increases, Profit tends to decrease.
#Red dashed horizontal line at Profit = 0
#Orders above this line made positive profit.
#Orders below this line caused a loss.

plt.figure(figsize=(8,5))
sns.scatterplot(data=tables, x='Discount', y='Profit', hue='Segment')
sns.regplot(data=tables, x='Discount', y='Profit', scatter=False, color='black')
plt.axhline(0, color='red', linestyle='--', label='Break-even Profit')
plt.title('Profit vs Discount (Tables)')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.show()
```

### Step7: impact of doscounts on profit

```python
import statsmodels.api as sm

```

```python
# Create discount bins
tables['Discount_bin'] = pd.cut(tables['Discount'], bins=[0, 0.1, 0.2, 0.3, 0.4, 1])

# Convert Interval to string for plotting
impact_df = tables.groupby('Discount_bin')['Profit'].mean().reset_index()
impact_df['Discount_bin'] = impact_df['Discount_bin'].astype(str)

# Plot line chart
plt.figure(figsize=(8,5))
sns.lineplot(data=impact_df, x='Discount_bin', y='Profit', marker='o')

plt.title('Average Profit by Discount Range (Tables)')
plt.xlabel('Discount Range')
plt.ylabel('Average Profit')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
tables['Discount_bin'] = pd.cut(tables['Discount'], bins=[0, 0.1, 0.2, 0.3, 0.4, 1])
order_counts = tables.groupby(['Segment', 'Discount_bin']).size().reset_index(name='Order_Count')
order_counts['Discount_bin'] = order_counts['Discount_bin'].astype(str)
```

```python

plt.figure(figsize=(10,6))
sns.barplot(data=order_counts, x='Discount_bin', y='Order_Count', hue='Segment')

plt.title('Number of Orders by Discount Range and Segment (Tables)')
plt.xlabel('Discount Range')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()
```
